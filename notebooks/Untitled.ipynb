```python
import pandas as pd

target = pd.read_csv('./../wip/MonthlySource.csv', parse_dates=['start'])

key_column ='source'
metric ='ava_debt'
target['start'] = target['start'].dt.strftime('%Y-%m')
monthly_pivot = target[target[metric] > 0].groupby(['source' ,'start']).agg(
    {metric: 'mean'}).reset_index().pivot(
    index='source' , columns='start', values=metric
).reset_index().fillna(0).reset_index()

monthly_pivot

```
