```python
import pandas as pd
import math
import statistics
```

```python
target = pd.read_csv('./../template/generate_data.csv', sep=';', names = ['source', 'start', 'end', 'total', 'ava', 'exp', 'lat'])
```

```python
target
```

```python
def avg_x(x):
    temp = [y for y in x if y > 0]
    return statistics.mean(temp)

metric = 'ava'
key_column = 'source'
monthly_pivot = target[target[metric] > 0].groupby([key_column, 'start']).agg(
            { metric: avg_x } ).reset_index().pivot(
            index=key_column, columns='start', values=metric
        ).reset_index().fillna(0).reset_index()
```
